```python


import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
# Count number of times each value in koi_disposition occurs
df["koi_disposition"].value_counts()
```

```python
df.columns
```

# Select your features (columns)

```python
X = df[["koi_fpflag_nt", "koi_fpflag_ss","koi_fpflag_co","koi_fpflag_ec","koi_period",
                "koi_time0bk","koi_impact","koi_duration","koi_prad","koi_teq","koi_insol", 
                     'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_steff_err1',
                    'koi_slogg', 'koi_srad',  'ra', 'dec',
                          'koi_kepmag']]
y = df["koi_disposition"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
data = X.copy()

```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

X = pd.get_dummies(data)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

X_train.head()
```

```python
y_train= y_train.ravel()
```

```python
y_train
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
#Transform training and testing data using Xscaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from tensorflow.keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
from tensorflow.keras.utils import to_categorical
y_train_categorical = to_categorical(one_hot_y )
y_test_categorical =to_categorical(one_hot_y )
y_train_categorical
```

# Train the Model



```python
# create the SVC model, try different kernels
# Create the SVC Model
from sklearn.svm import SVC 

model = SVC(kernel='linear')

model.fit(X_train_scaled,y_train)
```

```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
# param_grid = {'C': [1, 5, 10, 50],
#               'gamma': [0.0001, 0.0005, 0.001, 0.005]}
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.0005, 0.001]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python

# Train the model with GridSearch
grid.fit(X_train_scaled,y_train)
```

```python
# print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test_scaled)
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["CANDIDATE","FALSE POSITIVE","CONFIRMED"]))
```

## Analysis: 
   Precision  for CONFIRMED is 98% the other attributes(CANDIDATE,FALSE POSITIVE) has only 70% precision.so over all accuracy is 83%, need to check with other models.

## KNN Classifier

```python
# KNN classifier
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(X_train_scaled, y_train)
print('k=13 Test Acc: %.3f' % knn.score(X_test_scaled, y_test))
```

```python
predicted_class = knn.predict(X_test)
print(predicted_class)
```
